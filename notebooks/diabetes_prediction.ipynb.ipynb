```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from google.colab import files
uploaded = files.upload()
import pandas as pd

# Charger le fichier
df = pd.read_csv("diabetes.csv")

# Aperçu des données
df.head()
# Taille du dataset
print("Nombre de lignes et de colonnes :", df.shape)

# Types de données et valeurs nulles
print("\nInfos sur les colonnes :")
df.info()

# Statistiques descriptives
print("\nStatistiques :")
df.describe()
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols_with_zero:
    zero_count = (df[col] == 0).sum()
    print(f"{col} : {zero_count} zéros")
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(figsize=(12,10), bins=20)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
sns.boxplot(x="Outcome", y="Glucose", data=df)
plt.title("Glycémie vs Diabète")
plt.show()

sns.boxplot(x="Outcome", y="Age", data=df)
plt.title("Âge vs Diabète")
plt.show()

sns.boxplot(x="Outcome", y="BMI", data=df)
plt.title("IMC vs Diabète")
plt.show()
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Remplacement des 0 par NaN
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)

# Vérification
df.isnull().sum()
df.fillna(df.median(numeric_only=True), inplace=True)

# Vérif finale
df.isnull().sum()
# Features et target
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
print(df.head())
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)
print("Forme de X_train :", X_train.shape)
print("\nRépartition des classes dans y_train :")
print(y_train.value_counts())
from sklearn.linear_model import LogisticRegression
# Création du modèle
logreg = LogisticRegression(max_iter=1000)
# Entraînement
logreg.fit(X_train, y_train)
# Prédiction sur le test
y_pred_log = logreg.predict(X_test)
# Affichage des 10 premières prédictions
print("🔎 Premières prédictions :", y_pred_log[:10])
from sklearn.ensemble import RandomForestClassifier

# Modèle Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Prédiction
y_pred_rf = rf.predict(X_test)

# Affichage des premières prédictions
print("🔎 Prédictions Random Forest :", y_pred_rf[:10])
print("🔍 Logistic Regression")
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
print("Accuracy :", accuracy_score(y_test, y_pred_log))
print("🌲 Random Forest")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
fig, axs = plt.subplots(1, 2, figsize=(12,5))

# Logistic Regression
sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt="d", cmap="Blues", ax=axs[0])
axs[0].set_title("Logistic Regression")
axs[0].set_xlabel("Prédit")
axs[0].set_ylabel("Réel")

# Random Forest
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Greens", ax=axs[1])
axs[1].set_title("Random Forest")
axs[1].set_xlabel("Prédit")
axs[1].set_ylabel("Réel")

plt.tight_layout()
plt.show()




```
